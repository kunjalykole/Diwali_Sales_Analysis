```python
import pandas as pd # For data cleaning
import numpy as numpy # For numeric computational
import matplotlib.pyplot as plt # Data Visualization
import seaborn as sns # Data Visualization
```

```python
df=pd.read_csv("/content/Diwali Sales Data.csv", encoding='unicode_escape')
```

```python
df
```

```python
# Dropping null values
df.drop(columns=['unnamed1','Status'],inplace=True)
```

```python
# Renaming maritial status
df.rename(columns={'Marital_Status':'Vaivahik stithi'})
```

```python
df.info()
```

```python
# getting % of missing values
df.isnull().mean()*100
```

```python
# Dropped the missing values
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# astype used to convert float value into int
df['Amount']=df['Amount'].astype('int')
df
```

```python
df.describe()
```

```python
# Finding duplicated values
df.duplicated().sum()
```

```python
# Removing duplicated values using drop_duplicates
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

# EDA - Exploratory Data Analysis

```python
# plotting
df
```

```python
# countplot - to represent the occurrence(counts) of the observation present in the categorical variable.
sns.countplot(x="Gender", data=df)
```

```python
# plot a graph for gender vs total amount spent (reset_index converting series into dataframe)
sales_gen=df.groupby('Gender')['Amount'].sum().reset_index()
sales_gen
```

```python
sns.barplot(data=sales_gen,x='Gender',y='Amount')
```

```python
# plotting a graph for gender vs total amount spent(using index and values position)
```

```python
sales_gen=df.groupby('Gender')['Amount'].sum()
sales_gen
```

```python
sales_gen.index
```

```python
sales_gen.values
```

```python
sns.barplot(x=sales_gen.index,y=sales_gen.values)
```

# Documentation
From the above graph what insights we get
1. The purchasing power of female is more amount than male.

```python
# Total amount spent by each age, age group, state, occupation
```

```python
# Total amount spent by age group visualizing using pandas plot function
df['Age Group'].value_counts().plot(kind='bar')
```

```python
# Total highest amount spent by age group visualizing using pandas plot function and using palettes seaborn function
sales_ageGrp=df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
```

```python
sales_ageGrp.index
```

```python
sales_ageGrp.values
```

```python
sns.barplot(x=sales_ageGrp.index,y=sales_ageGrp.values,palette='viridis')
```

# Documentation
From the above graph we understand that from age group 26-35 spent more amount of money.

```python
# Total amount spent by age group visualizing using seaborn and segregating based on male and female
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
# Plotting graph of states based on their order's total quantity
```

```python
# Plotting graph of states based on their amount spent
```

```python
statesAmt=df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
```

```python
statesAmt.index
```

```python
statesAmt.values
```

```python
sns.barplot(x=statesAmt.index,y=statesAmt.values,palette='viridis')
plt.xticks(rotation=90)
```

# Documentation
From above graphs we can observe that most of the orders and highest amount spent are from UP, MH, Karnataka

```python
# Plotting graph of amount spent by each occupation
```

```python
statesOcc=df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).head(10)
```

```python
statesOcc.index
```

```python
statesOcc.values
```

```python
sns.barplot(x=statesOcc.index,y=statesOcc.values,palette='pastel')
plt.xticks(rotation=90)
```

```python
marStatus=df.groupby(['Marital_Status','Gender'])['Amount'].sum().reset_index() # used reset_index to convert x axis series in dataframe
marStatus
```

# Documentation
By above graph we can observe that in unmarried and married people females have spent more money. We also observed that unmarried people have spent more than married.
